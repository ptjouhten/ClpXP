```python
#introduce heterologous muconic acid synthesis pathway in yeast
reaction = Reaction('AnOXA')
reaction.name = 'Oxaloacetase'
reaction.subsystem = 'oxalate pathway'
reaction.lower_bound = 0.  # This is the default
reaction.upper_bound = 1000.  # This is the default
```

```python
oxalate_c = Metabolite(
    'oxalate_c',
    formula='C2H2O4',
    name='oxalate',
    compartment='c')
oxalylcoa_c = Metabolite(
    'oxalylcoa_c',
    formula='C23H36N7O19P3S',
    name='oxalylcoa',
    compartment='c')
glycolic_acid_c = Metabolite(
    'glycolic_acid_c',
    formula='C2H4O3',
    name='glycolic_acid',
    compartment='c')
glycolic_acid_e = Metabolite(
    'glycolic_acid_e',
    formula='C2H4O3',
    name='glycolic_acid',
    compartment='e')
```

```python
reaction.add_metabolites({
    oxa: -1.0,
    #h2o: -1.0,
    oxalate_c: 1.0,
    #acetate: 1.0,
    #h: 1.0
})

reaction.reaction 
```

```python
Sc.add_reactions([reaction])
```
